```python
plt.style.use('_mpl-gallery')
 
# code to plot 
plt.figure(figsize=(10,6))
plt.hist(df['Species'], color='#40e0d0',width = 0.3, label='Frequency of species')
# title of the plot
plt.title("Histogram of the count of species")
# add labels to the axis
plt.xlabel("Species name")
plt.ylabel('Count of species')
# grid
# legend
plt.legend()
plt.show()
```

```python
plt.style.use('_mpl-gallery')
 
# code to plot 
plt.figure(figsize=(10,6))
plt.scatter('SepalLengthCm','SepalWidthCm',marker='o', s=35, linestyle='-',data=df)
# title of the plot
plt.title("A scatter plot with iris data", fontsize=14,color='Blue' )
# add labels to the axis
plt.xlabel("SepalLengthCm", fontsize=14)
plt.ylabel('SepalWidthCm', fontsize=14)
# grid
# legend
plt.legend(loc="upper left",title="Sepal width and length", title_fontsize=12)
plt.show()
```

```python
data_plot_group=df.groupby('Species').agg({'SepalLengthCm':'mean','SepalWidthCm':'mean',
                                           'PetalLengthCm':'mean','PetalWidthCm':'mean'}).reset_index()
data_plot_group
```

```python
plt.style.use('_mpl-gallery')
 
# code to plot 
plt.figure(figsize=(10,6))
plt.bar(data_plot_group.Species,data_plot_group.SepalLengthCm, color='#40e0d0',width = 0.3,label='Species')
# title of the plot
plt.title("A bar plot with average Sepal Length per species",fontsize=16,color='Green')
# add labels to the axis
plt.xlabel("Species",fontsize=12)
plt.xticks(rotation=0, fontsize=10)
plt.ylabel('Average sepal length',fontsize=12)
plt.yticks(fontsize=12)
# grid
# legend
plt.legend()
plt.show()
```
