```python
country_sorted=df_sales['COUNTRY'].value_counts().sort_values().reset_index

sns.set_theme(style='whitegrid')
plt.figure(figsize=(16,6))
plt.bar('COUNTRY','count', color='#40e0d0',width = 0.5, label='Value over Years', data=country_sorted)
plt.xticks(rotation=45)
plt.legend()
plt.show()
plt.clf()
```

```python
country_sorted=df_sales['COUNTRY'].value_counts().sort_values()

plt.style.use('_mpl-gallery')
 
# code to plot 
plt.figure(figsize=(10,6))
plt.bar('COUNTRY','count', color='#40e0d0',width = 0.5, label='Value over Years', data=country_sorted)
# title of the plot
plt.title("A bar plot with fictional data")
# add labels to the axis
plt.xlabel("Years")
plt.xticks(rotation=45)
plt.ylabel('Sales of products (in thousands)')
# grid
# legend
plt.legend()
plt.show()
```
